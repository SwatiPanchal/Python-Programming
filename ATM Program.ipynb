```python
import random
import datetime
import numpy as np

def create_pin(pin):
    data[pin] = {'Date' : [], 'Transaction Type' : [], 'Deposit' : [], 'Withdraw' : [], 'Balance' : []}
    # Saving the data
    np.save('database.npy',data)
  
def update_database(amount,t,p):
    data[p]['Date'].append(datetime.datetime.now())
    if t=="d":
        data[p]['Transaction Type'].append("Deposit")
        data[p]['Deposit'].append(amount)
        data[p]['Withdraw'].append(0)
        if len(data[p]['Balance'])>0:
            data[p]['Balance'].append(data[p]['Balance'][-1]+amount)
        else:
            data[p]['Balance'].append(amount)
    else:
        data[p]['Transaction Type'].append("Withdraw")
        data[p]['Deposit'].append(0)
        data[p]['Withdraw'].append(amount)
        data[p]['Balance'].append(data[p]['Balance'][-1]-amount)
        
    # Saving the data
    np.save('database.npy',data)


def activate_pin():
    pin = int(input("\nEnter a 4-digit pin number :"))
    if len(str(pin)) == 4:
        if pin not in data:
            randnum = random.randint(1111,9999)
            print("Your otp for pin activation is {}".format(randnum))
            OTP_Verification(randnum)
            create_pin(pin)
        else:
            print("\nPin already exists in database")
            print("Please select another pin number")
            activate_pin()
    else:
        print("Enter valid pin number")
        pin = input("Enter a 4-digit pin number :")
        activate_pin()

        
def OTP_Verification(randnum):
    otp = int(input("Enter the OTP for verification :"))
    if randnum == otp:
        print("Your pin has been activated\n")
    else:
        print("OTP is incorrect. Enter valid OTP")
        OTP_Verification(randnum)

        
def change_pin():
    pin = int(input("Plaese enter the old pin : "))
    if pin in data:
        new_pin = int(input("Please enter the new pin : "))
        re_enter = int(input("Please re-enter your new pin : "))
        if new_pin==re_enter:
            randnum = random.randint(1111,9999)
            print("Your otp for pin activation is {}".format(randnum))
            OTP_Verification(randnum)
            data[new_pin] = data[pin]
            del data[pin]
        else:
            print("\nYou entered wrong pin")
            print("Please do the process again\n")
    else:
        print("\nYou entered wrong pin")
        print("Please do the process again\n")
            
def deposit():
    
    amount = int(input("\nEnter the amount to deposit :"))
    pin_num = int(input("Enter your 4-digit pin number :"))
    if pin_num in data:
        update_database(amount,"d",pin_num)
        print("\nYour amount {} has been deposited".format(amount))
        display(pin_num)
    else:
        print("Enter Valid pin")
        print("Your Transaction has been cancelled \n")


def withdraw(data):
    amount = int(input("\nEnter the amount to withdraw :"))
    pin_num = int(input("Enter your 4-digit pin number :"))
    if pin_num in data:
        if len(data[pin_num]['Balance'])>=1:
            if (data[pin_num]['Balance'][-1]-amount)>0:
                data = update_database(amount,"w",pin_num)
                print("Your amount {} has been withdrawn".format(amount))
                display(pin_num)
            else:
                print("You do not have sufficient balance.")
                display(pin_num)
        else:
            print("You do not have sufficient balance.")
            display(pin_num)
            
    else:
        print("Enter Valid pin")
        print("Your Transaction has been cancelled \n")

        
def display(pin):
    if len(data[pin]['Balance'])>=1:
        print("Your Available Balance = {} \n".format(data[pin]['Balance'][-1]))
    else:
        print("Your Available Balance = 0")

    print("Do you want mini statement : y/n")
    ans = input()
    if ans == "y" or ans =="Y":
        mini_statement(pin)
    elif ans == 'n' or ans == 'N':
        pass
    else:
        print("Sorry!! You entered wrong choice")
        

def mini_statement(pin):
    
    print()
    if len(data[pin]['Date'])>=1:
        print(data[pin]['Date'][-1])
        if data[pin]['Transaction Type'][-1] == 'Deposit':
            print('Deposit'.ljust(20), ":", data[pin]['Deposit'][-1])
        else:
            print("Withdraw Amount".ljust(20) , ":", data[pin]['Withdraw'][-1])
        if len(data[pin]['Balance'])<=1:
            print("Regular Balance".ljust(20), ":", 0)
        else:
            print("Regular Balance".ljust(20), ":", data[pin]["Balance"][-2])

        print("Availavble Balance".ljust(20), ":", data[pin]['Balance'][-1])
    else:
        print("You have not done any transaction yet")
        print("Your available balance = 0")


def balance_enquiry():

    pin_num = int(input("\nEnter your 4-digit pin number to check your balance :"))
    if pin_num in data:
        if len(data[pin_num]['Balance'])>0:
            print("\nYour Available balance is Rs.{}".format(data[pin_num]['Balance'][-1]))
        else:
            print("\nYour Available balance is Rs.0")
    else:
        print("\nEnter Valid pin")
        print("Your Transaction has been cancelled \n")



def account_stat():
            
    pin = int(input("Enter your pin number :"))
    if pin in data:
        print("\nState Bank Of India \n")
        if len(data[pin]["Date"])>=1:
            for j in range(len(data[pin]['Date'])):
                for k in data[pin]: 
                    print(k.ljust(20), ":" ,data[pin][k][j])
                print()
        else:
            print("You don't have any transactions")
    else:
        print("Enter valid pin number")
        
print("State Bank Of India \n")
print("Activate Your PIN")

# Loading the data
def load_database():
    data = np.load('database.npy', allow_pickle=True).item()
    return data

data = load_database()

while True:
    print()
    print("1) Activate Pin")
    print("2) Deposit Money")
    print("3) Withraw Money")
    print("4) Balance Enquiry")
    print("5) Account Satement")
    print("6) Change Pin")
    print("7) Exit")
    print()

    choice = int(input("Enter your choice :"))
    if choice==1:
        activate_pin()
    elif choice==2:
        deposit()
    elif choice==3:
        withdraw(data)
    elif choice==4:
        balance_enquiry()
    elif choice==5:
        account_stat()
    elif choice==6:
        pin = change_pin()
    elif choice==7:
        print("\nThank You For Using SBI ATM")
        break
    else:
        print("\nEnter valid choice")

```
